# Autoencoder

Why to use all this complicated formulaes and regularizations, what is the need for variational inference? To analyze the difference, let's first train just an autoencoder on the data:

<img src="Autoencoder_structure.png" alt="Autoencoder">

```python
dimZ = 100 # Considering face reconstruction task, which size of representation seems reasonable?

# Define the decoder and encoder as networks with one hidden fc-layer
# (that means you will have 2 fc layers in each net)
# Use ReLU for hidden layers' activations
# GlorotUniform initialization for W
# Zero initialization for biases
# It's also convenient to put sigmoid activation on output layer to get nice normalized pics

class Autoencoder(nn.Module):
    def __init__(self):
        super(Autoencoder, self).__init__()
        layers_enc = [
            nn.Linear(input_dim, 1024),
            nn.ReLU(),
            nn.Linear(1024,dimZ),
            nn.Sigmoid()#from TA
        ]
        
        layers_dec = [
            nn.Linear(dimZ,1024),
            nn.ReLU(),
            nn.Linear(1024,input_dim),
            nn.Sigmoid()#from TA
        ]

        # https://discuss.pytorch.org/t/how-are-layer-weights-and-biases-initialized-by-default/13073
        def weights_init(m):
            if isinstance(m,nn.Linear):
                torch.nn.init.xavier_uniform_(m.weight)
                torch.nn.init.constant_(m.bias,0.0)

        self.encoder = nn.Sequential(*layers_enc)
        self.encoder.apply(weights_init)
        self.decoder = nn.Sequential(*layers_dec)
        self.decoder.apply(weights_init)
        
    def forward(self, x):
        
        latent_code = self.encoder(x)
        reconstruction = self.decoder(latent_code)
        
        return reconstruction, latent_code
```

```python
# Create MSE loss function
criterion = torch.nn.MSELoss()

autoencoder = Autoencoder().cuda()

# Use Adam optimizer
optimizer = optim.Adam(autoencoder.parameters())
```

```python
# Train your autoencoder
# Visualize progress in reconstruction and loss decay
import time
from tqdm import tqdm
from IPython import display

batch_size = 100


def train_ae(model, opt=optimizer, num_epochs=15,criterion=criterion, Putin=True):

    train_loss = []
    val_loss = []
    
    for i, (putin, y) in enumerate(val_loader):
        if i == 2754:
            break
    if Putin:
        image_progress = [putin]
      
    
    

    for epoch in tqdm(range(num_epochs)):
        # In each epoch, we do a full pass over the training data:
        start_time = time.time()
        model.train(True) # enable dropout / batch_norm training behavior
        for (X_batch, y_batch) in train_loader:
            # train on batch
            X_batch = Variable(X_batch).cuda()
            recon = model(X_batch)[0]
            loss = criterion(recon,X_batch)
            loss.backward()
            opt.step()
            opt.zero_grad()
            train_loss.append(loss.data.cpu().numpy()) 
            
        model.train(False) # disable dropout / use averages for batch_norm
        inp = Variable(putin.cuda())
        pred = model(inp)[0]
        image_progress.append(pred)
        for i ,(X_batch, y_batch) in enumerate(val_loader):
            X_batch = Variable(X_batch).cuda()
#             recon =  model(X_batch)[0]
#             if Putin:
#                 if i == 2754:
#                     print(2)
# #               plt.imshow(recon.numpy().reshape((image_w, image_w, 3)))
#                     image_progress.append(recon)
            loss = criterion(recon,X_batch)
            val_loss.append(loss.data.cpu().numpy())
        # Then we print the results for this epoch:
        print("Epoch {} of {} took {:.3f}s".format(epoch + 1, num_epochs, time.time() - start_time))
        print("  training loss (in-iteration): \t{:.6f}".format(np.mean(train_loss[-len(train_loader) // batch_size :])))
        print("  validation loss (in-iteration): \t{:.6f}".format(np.mean(val_loss[-len(val_loader) :])))
        
    # Visualize loss
    plt.figure(figsize=(10, 6))

    plt.title("loss decay")
    plt.xlabel("iteration")
    plt.ylabel("loss")
    plt.semilogy(train_loss, label = 'training loss')
#     plt.semilogy(val_loss, label = 'validation loss')
    plt.show()
    if Putin:
        return image_progress
```

# Variational Autoencoder

Bayesian approach in deep learning considers everything in terms of distributions. Now our encoder generates not just a vector $z$ but posterior ditribution $q(z|x)$. In our case distribution $q$ is Gaussian distibution $N(\mu, \sigma)$ with parameters $\mu$, $\sigma$. Technically, the first difference is that you need to split bottleneck layer in two. One dense layer will generate vector $\mu$, and another will generate vector $\sigma$. Reparametrization trick should be implemented using the **gaussian_sampler** layer, that generates random vetor $\epsilon$ and returns $\mu+\sigma\epsilon \sim N(\mu, \sigma)$ .

Since our decoder is also a function that generates distribution, we need to do the same splitting for output layer. When testing the model we will look only on mean values, so one of the output will be actual autoencoder output.

In this homework we only ask for implementation of the simplest version of VAE - one $z$ sample per input. You can consider to sample several outputs from one input and average them.

# it looks like in https://github.com/pytorch/examples/blob/master/vae/main.py

```python
# to compare with conventional AE, keep these hyperparameters
# or change them for the values that you used before
dimZ = 100

# define the network
# you can start from https://github.com/pytorch/examples/blob/master/vae/main.py
# or Theano-based examples here https://github.com/Lasagne/Recipes/blob/master/examples/variational_autoencoder/variational_autoencoder.py
# and here https://github.com/y0ast/Variational-Autoencoder/blob/master/VAE.py
# but remember that this is not your ground truth since the data is not MNIST

class VAE(nn.Module):
    def __init__(self):
        super(VAE, self).__init__()
        
        self.fc1 = nn.Linear(input_dim, 1024)
    
        self.fc21 = nn.Linear(1024,dimZ)
        self.fc22 = nn.Linear(1024,dimZ)
        self.fc3 = nn.Linear(dimZ,1024)
        self.fc41 =nn.Linear(1024,input_dim)
        self.fc42 =nn.Linear(1024,input_dim)
        
        
    def encoder(self,x):
        h1 = F.relu(self.fc1(x))
        return self.fc21(h1),self.fc22(h1)
    
    def decoder(self, z):
        h3 = F.relu(self.fc3(z))
#         h3 = self.relu(self.bn2(self.fc3(z)))
        return torch.sigmoid(self.fc41(h3)),torch.sigmoid(self.fc42(h3))
    
    
    def gaussian_sampler(self, mu, logsigma):
        if self.training:
            std = logsigma.exp_()
            eps = Variable(std.data.new(std.size()).normal_())
            return eps.mul(std).add_(mu)
        else:
            return mu

    def forward(self, x):
        mu, logvar = self.encoder(x)
        z = self.gaussian_sampler(mu, logvar) 
        rec_mu, rec_logsigma = self.decoder(z) 
        return rec_mu, rec_logsigma , mu, logvar
```

And the last, but not least! Place in the code where the most of the formulaes goes to - optimization objective. The objective for VAE has it's own name - variational lowerbound. And as for any lowerbound our intention is to maximize it. Here it is (for one sample $z$ per input $x$):

$$\mathcal{L} = -D_{KL}(q_{\phi}(z|x)||p_{\theta}(z)) + \log p_{\theta}(x|z)$$

Your next task is to implement two functions that compute KL-divergence and the second term - log-likelihood of an output. Here is some necessary math for your convenience:

$$D_{KL} = -\frac{1}{2}\sum_{i=1}^{dimZ}(1+log(\sigma_i^2)-\mu_i^2-\sigma_i^2)$$
$$\log p_{\theta}(x|z) = \sum_{i=1}^{dimX}\log p_{\theta}(x_i|z)=\sum_{i=1}^{dimX} \log \Big( \frac{1}{\sigma_i\sqrt{2\pi}}e^{-\frac{(\mu_I-x)^2}{2\sigma_i^2}} \Big)=...$$

Don't forget in the code that you are using $\log\sigma$ as variable. Explain, why not $\sigma$?

И последнее, но не менее важное! Место в коде, куда идет большинство формул - цель оптимизации. Цель VAE имеет собственное имя - вариационная нижняя граница. И что касается любого нижнего предела, мы намерены максимизировать его. Вот оно (для одной выборки $ z $ на вход $ x $):

$$\mathcal{L} = -D_{KL}(q_{\phi}(z|x)||p_{\theta}(z)) + \log p_{\theta}(x|z)$$

Ваша следующая задача - реализовать две функции, которые вычисляют KL-расхождение, и второе слагаемое - логарифмическая вероятность вывода. Вот некоторая необходимая математика для вашего удобства:

$$D_{KL} = -\frac{1}{2}\sum_{i=1}^{dimZ}(1+log(\sigma_i^2)-\mu_i^2-\sigma_i^2)$$
$$\log p_{\theta}(x|z) = \sum_{i=1}^{dimX}\log p_{\theta}(x_i|z)=\sum_{i=1}^{dimX} \log \Big( \frac{1}{\sigma_i\sqrt{2\pi}}e^{-\frac{(\mu_I-x)^2}{2\sigma_i^2}} \Big)=...$$

Не забудьте в коде, что вы используете $ \ log \ sigma $ в качестве переменной. Объясните, почему не $ \ sigma $?

```python
def KL_divergence(mu, logsigma):
    return -0.5*torch.sum(1+2*logsigma-mu*mu - torch.exp(2*logsigma))
     

def log_likelihood(x, mu, logsigma):
    return torch.sum( -logsigma -0.5*np.log(2 *np.pi) - ((mu - x)*(mu-x))/(2 * torch.exp(2 * logsigma)))
 

def loss_vae(x, mu_gen, logsigma_gen, mu_z, logsigma_z):
    return -(-KL_divergence(mu_z, logsigma_z)+log_likelihood(x, mu_gen, logsigma_gen))
```

And train the model:

```python
# train your autoencoder
# visualize progress in reconstruction and loss decay
def train_vae(model, opt=optimizer, num_epochs=15,criterion=loss_vae):

    train_loss = []
    val_loss = []
    
    for epoch in tqdm(range(num_epochs)):
        # In each epoch, we do a full pass over the training data:
        start_time = time.time()
        model.train(True) # enable dropout / batch_norm training behavior
        for (X_batch, y_batch) in train_loader:
            # train on batch
            X_batch = Variable(X_batch).cuda()
            rec_mu, rec_logsigma, lat_mu, lat_logsigma = model(X_batch)

            loss = loss_vae(x=X_batch,
                            mu_gen=rec_mu,
                            logsigma_gen=rec_logsigma,
                            mu_z=lat_mu,
                            logsigma_z=lat_logsigma)
            loss.backward()
            opt.step()
            opt.zero_grad()
            train_loss.append(loss.data.cpu().numpy()) 
            

        print("Epoch {} of {} took {:.3f}s".format(epoch + 1, num_epochs, time.time() - start_time))
        print("  training loss (in-iteration): \t{:.6f}".format(np.mean(train_loss[-len(train_loader) // batch_size :])))
        # Visualize loss
    plt.figure(figsize=(10, 6))

    plt.title("loss decay")
    plt.xlabel("iteration")
    plt.ylabel("loss")
    plt.semilogy(train_loss[10:], label = 'training loss')
#     plt.semilogy(val_loss, label = 'validation loss')
    plt.show()
       
```

```python
vae = VAE().cuda()
# Use Adam optimizer
optimizer = optim.Adam(vae.parameters())

train_vae(vae,opt=optimizer,criterion=loss_vae,num_epochs=70)
```
